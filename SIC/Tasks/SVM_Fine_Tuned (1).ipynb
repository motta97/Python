```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import numpy as np

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['Gaussian','rbf', 'linear','polynomial']
}

grid = GridSearchCV(
    SVC(),
    param_grid,
    refit=True,
    cv=2,
    scoring='accuracy'
)

nan_indices = np.isnan(x_train).any(axis=1)
x_train_cleaned = x_train[~nan_indices]
y_train_cleaned = y_train[~nan_indices]

grid.fit(x_train_cleaned, y_train_cleaned)
```

```python
grid.best_params_
```

Now that the model is trained, let's evaluate its performance on the test set.

```python
from sklearn.metrics import accuracy_score
nan_indices_test = np.isnan(x_test).any(axis=1)
x_test_cleaned = x_test[~nan_indices_test]
y_test_cleaned = y_test[~nan_indices_test]
y_pred = grid.predict(x_test_cleaned)
accuracy_score(y_test_cleaned, y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
import numpy as np

param_dist = {
    'C': np.logspace(-2, 2, 50),
    'gamma': np.logspace(-4, 1, 50),
    'kernel': ['rbf', 'linear']
}

random_search = RandomizedSearchCV(
    SVC(),
    param_distributions=param_dist,
    n_iter=20,
    cv=3,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)


random_search.fit(x_train_cleaned, y_train_cleaned)

print("Best Parameters:", random_search.best_params_)
print("Best CV Accuracy:", random_search.best_score_)
```

```python
y_pred_random=random_search.predict(x_test_cleaned)
```

```python
accuracy_score(y_pred_random,y_test_cleaned)
```
