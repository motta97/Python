```python
import numpy as np
```

```python
A = np.array([[1,4,3],[6,5,0],[1,9,0]])
B = np.array([[9,8,7],[6,5,4],[3,2,1]])
```

```python
A+B*A+B*A+B
```

```python
A-B*A-B*A-B
```

```python
A.T
```

```python
np.linalg.det(A)
```

```python
A_INV=np.linalg.inv(A)
```

```python
A_INV
```

```python
A@A_INV
```
