```python
# we will also do as before and categorize them since we dont really care about delay in numbers
# firstly we will need to impute the missing values in Arrival Delay in Minutes
# we will use Median imputation for arrival delay → simple, robust to outliers.
# import knearst neighbors imputer
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
df['Arrival Delay in Minutes'] = imputer.fit_transform(df[['Arrival Delay in Minutes']])

def bucket_delay(minutes):
    if minutes == 0:
        return "On time"
    elif 1 <= minutes <= 30:
        return "Slight delay"
    elif 31 <= minutes <= 120:
        return "Moderate delay"
    else:
        return "Severe delay"

# Apply for both columns
df["Departure Delay in Minutes"] = df["Departure Delay in Minutes"].apply(bucket_delay)
df["Arrival Delay in Minutes"] = df["Arrival Delay in Minutes"].apply(bucket_delay)

# Convert to category dtype
df["Departure Delay in Minutes"]  = df["Departure Delay in Minutes"].astype("category")
df["Arrival Delay in Minutes"]   = df["Arrival Delay in Minutes"].astype("category")
```

## Feature encoding

```python
df.info()
```

```python
# for categorical cols of <= 2 unique values -> we will do label encoding
# for categorical cols of > 2 unique values -> we will do one-hot encoding
from sklearn.preprocessing import LabelEncoder

data_copy = df.copy()

# Find categorical columns
cat_cols = df.select_dtypes(['category']).columns

# Loop through categorical columns
for col in cat_cols:
    if df[col].nunique() <= 2:
        le = LabelEncoder()
        data_copy[col] = le.fit_transform(df[col])
    else:
        dummies = pd.get_dummies(df[col], prefix=col, drop_first=True)
        data_copy = pd.concat([data_copy.drop(col, axis=1), dummies], axis=1)
```

# **Modelling**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
import lightgbm as lgb
```

```python
test_data = pd.read_csv("./test.csv")
test_data_copy = test_data.copy()
# preprocessing the test data the same way as train data
test_data = pd.read_csv("./test.csv") 
test_data_copy = test_data.copy()

def preprocess(df):
    df = df.copy()
    
    df.drop(columns=['Unnamed: 0', 'id'], inplace=True, errors='ignore')
    
    # 2. Convert categorical cols
    if 'Gender' in df:
        df['Gender'] = df['Gender'].astype("category")
    if 'Customer Type' in df:
        df['Customer Type'] = df['Customer Type'].astype("category")
    if 'Type of Travel' in df:
        df['Type of Travel'] = df['Type of Travel'].astype("category")
    if 'Class' in df:
        df['Class'] = pd.Categorical(df['Class'], categories=['Eco', 'Eco Plus', 'Business'])
    
    # 3. Flight Distance binning
    if 'Flight Distance' in df:
        bins = [0, 1500, 3000, df['Flight Distance'].max()]  
        labels = ['Short-haul', 'Medium-haul', 'Long-haul']
        df['Flight Distance'] = pd.cut(df['Flight Distance'], bins=bins, labels=labels, include_lowest=True)
        df['Flight Distance'] = df['Flight Distance'].astype("category")
    
    # 4. Handle Delay Columns
    if 'Arrival Delay in Minutes' in df:
        imputer = KNNImputer(n_neighbors=5)
        df['Arrival Delay in Minutes'] = imputer.fit_transform(df[['Arrival Delay in Minutes']])
        
        def bucket_delay(minutes):
            if minutes == 0:
                return "On time"
            elif 1 <= minutes <= 30:
                return "Slight delay"
            elif 31 <= minutes <= 120:
                return "Moderate delay"
            else:
                return "Severe delay"
        
        df["Departure Delay in Minutes"] = df["Departure Delay in Minutes"].apply(bucket_delay)
        df["Arrival Delay in Minutes"]   = df["Arrival Delay in Minutes"].apply(bucket_delay)
        df["Departure Delay in Minutes"] = df["Departure Delay in Minutes"].astype("category")
        df["Arrival Delay in Minutes"]   = df["Arrival Delay in Minutes"].astype("category")
    
    # 5. Satisfaction column
    if 'satisfaction' in df:
        df['satisfaction'] = df['satisfaction'].astype("category")
    
    # 6. Encoding
    for col in df.select_dtypes(['category']).columns:
        if df[col].nunique() <= 2:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col])
        else:
            dummies = pd.get_dummies(df[col], prefix=col, drop_first=True)
            df = pd.concat([df.drop(col, axis=1), dummies], axis=1)
    
    return df

test_data_copy = preprocess(test_data)

print("Number of columns in test data:",len(test_data_copy.columns))
print("Number of columns in train data:",len(data_copy.columns))
```

## Classification

```python
X_train = data_copy.drop("satisfaction", axis=1)
y_train = data_copy["satisfaction"]
X_test = test_data_copy.drop("satisfaction", axis=1)
y_test = test_data_copy["satisfaction"]
```

### Baseline model (Logistic regression)

```python
log_reg_pipeline = Pipeline([
    ('scaler', StandardScaler()),           
    ('logreg', LogisticRegression(max_iter=1000,class_weight='balanced',random_state=42))  
])


log_reg_pipeline.fit(X_train, y_train)
y_pred_log = log_reg_pipeline.predict(X_test)
y_proba_log = log_reg_pipeline.predict_proba(X_test)[:,1]

print("Logistic Regression Results")
print(classification_report(y_test, y_pred_log))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_log))
```

### Random forest

```python
rand_forest = RandomForestClassifier(n_estimators=200, random_state=42, class_weight="balanced")

rand_forest.fit(X_train, y_train)
y_pred_rf = rand_forest.predict(X_test)
y_proba_rf = rand_forest.predict_proba(X_test)[:,1]

print("Random Forest Results")
print(classification_report(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_rf))
```

```python
# lets check if model is overfitting by performing cross validation

scores = cross_val_score(rand_forest, X_train, y_train, cv=5, scoring='accuracy')
print("CV scores:", scores)
print("Mean accuracy:", scores.mean())
print("Std deviation:", scores.std())
```

```python
# Feature importance according to random forest

importances = rand_forest.feature_importances_
feat_importances = pd.Series(importances, index=X_train.columns)

# Sort and plot
plt.figure(figsize=(10,6))
sns.barplot(x=feat_importances.sort_values(ascending=False)[:15],
            y=feat_importances.sort_values(ascending=False).index[:15],
            palette="viridis")
plt.title("Top 15 Feature Importances - Random Forest")
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, f1_score, accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import GridSearchCV, ShuffleSplit

# Define pipeline (scaler + classifier)
rf_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier(random_state=42, n_estimators=200))
])

# Parameter grid (prefix with 'rf__' to target RF inside pipeline)

parameters = {
    'rf__max_features': [None, 'sqrt', 'log2'],
    'rf__min_samples_split': [2, 5, 10],
    'rf__min_samples_leaf': [1, 2, 4],
    'rf__class_weight': ['balanced', 'balanced_subsample', None]
}


# Scorer
scorer = make_scorer(f1_score)

# GridSearch with ShuffleSplit CV
grid_obj = GridSearchCV(
    rf_pipeline, 
    parameters, 
    scoring=scorer, 
    cv=ShuffleSplit(n_splits=5, test_size=0.2, random_state=42),
    n_jobs=-1,
    verbose=2
)


grid_fit = grid_obj.fit(X_train, y_train)

# Get best estimator
best_clf = grid_fit.best_estimator_

# Predictions
best_train_predictions = best_clf.predict(X_train)
best_test_predictions = best_clf.predict(X_test)


print('The training F1 Score is', f1_score(y_train, best_train_predictions))
print('The testing F1 Score is', f1_score(y_test, best_test_predictions))

train_accuracy = accuracy_score(y_train, best_train_predictions)
test_accuracy = accuracy_score(y_test, best_test_predictions)
print('The training accuracy is', train_accuracy)
print('The test accuracy is', test_accuracy)
best_clf
```
